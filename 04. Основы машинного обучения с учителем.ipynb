# Основы машинного обучения с учителем
Цель - безошибочные прогнозы на основе имеющихся данных.

Модель с точки зрения пользователя - это черный ящик
+ на вход подали данные: вектор признаков (features)
+ на выходе получили предсказания: $y$



Два этапа:
+ обучение
+ выполнение вывода


**Этап обучения**
+ показываем модели, какой должен быть выходной сигнал $y$ при заданном входном сигнале $x$
+ выданное моделью предсказание $\hat y$ сранивается с $y$ - в случае расхождения  обновляем модель так, чтобы уменьшить расхождение

**Этап выполнения вывода**
+ обученной модели показываем новые входные признаки, которые она не видела. 
+ адекватные модели обладают способностюь к *обобщению* (generalization)
